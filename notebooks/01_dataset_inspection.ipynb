### Label Encoding for Machine Learning

Machine learning models require numerical labels rather than text.  
We map the original mental-health categories to numerical anxiety levels:

- 0 → Low Anxiety  
- 1 → Moderate Anxiety  
- 2 → High Anxiety

This ensures the model can process the labels correctly during training.

```python
# Define label mapping from text to numerical anxiety levels
label_mapping = {
    'Normal': 0,
    'Stress': 1,
    'Anxiety': 1,
    'Depression': 2,
    'Suicidal': 2,
    'Bipolar': 2,
    'Personality disorder': 2
}

# Apply mapping to create a new column 'anxiety_level'
df['anxiety_level'] = df['status'].map(label_mapping)

# Check the first few rows to verify
df.head()
```
